```python
#creation of sample matrix
x_range = range(-180, -80, 20)
y_range = range(5, 30, 5)

ma = np.zeros(shape=(6,6))
ma[0,1:] = x_range
ma[1:,0] = y_range
ma[0,0] = np.nan

# test function:
def test_func(x, y):
    z = (x + y) / 10

    return z  

#looping
for y in range(1, len(ma[0,:])):
    for x in range(1, len(ma[:,0])):
        #print ma[0, x], ma[y, 0]
        ma[x, y] = test_func(ma[0, x], ma[y, 0])

ma
```
